```python
import openai


conversation_length = 0
max_conversations = 5
message_history = [
    {"role": "system", "content": "You are an AI tutor that assists school students with math homework problems. Do not reveal the correct answer."}
]


print("AI Tutor is ready! Type 'exit' to stop.")

while conversation_length < max_conversations:

    user_input = input("You: ")


    if "exit" in user_input.lower():
        print("AI Tutor: Exiting the program!")
        break


    message_history.append({"role": "user", "content": user_input})


    response = openai.chat.completions.create(
        model="gpt-3.5-turbo-16k",
        messages=message_history
    )


    chat_response = response.choices[0].message.content


    print("\nAI Tutor:")
    print(chat_response)
    print("\n")


    message_history.append({"role": "assistant", "content": chat_response})


    conversation_length = conversation_length + 1
```

```python
import openai

conversation_length = 0
max_conversations = 10

system_prompt = """You are an expert conversational travel itinerary maker.
Your goal is to design customized travel itineraries.
However, before you create the itinerary, you MUST know these 4 specific constraints from the user:
1. Destination
2. Budget
3. Number of people traveling
4. Duration of the trip (Number of days)

If the user does not provide all 4 of these details in their messages, ask them friendly follow-up questions to gather the missing information.
Do NOT generate any itinerary until you have all 4 constraints.
Once you have successfully gathered all 4 pieces of information, output a detailed, practical, and highly customized day-by-day itinerary."""

message_history = [
    {"role": "system", "content": system_prompt}
]

print("AI Travel Agent: Hi there! I'm your AI travel planner. Where are you dreaming of going? (Type 'exit' to stop)\n")

while conversation_length < max_conversations:
    user_input = input("You: ")

    if "exit" in user_input.lower():
        print("\nAI Travel Agent: Have a safe trip! Exiting program.")
        break

    message_history.append({"role": "user", "content": user_input})

    response = openai.chat.completions.create(
        model="gpt-3.5-turbo-16k",
        messages=message_history,
        temperature=0.7
    )

    chat_response = response.choices[0].message.content

    print("\nAI Travel Agent:")
    print(chat_response)
    print("\n" + "-"*50 + "\n")

    message_history.append({"role": "assistant", "content": chat_response})

    conversation_length += 1
```
